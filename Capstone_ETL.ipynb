```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import stats
import seaborn as sns; sns.set()
import warnings
warnings.filterwarnings("ignore")
from datetime import datetime

tn = pd.read_csv('../data/Medicare_Provider_Charge_Inpatient_DRGALL_FY2015.csv')
```

```python
tn.columns = ['DRG', 'Provider_ID', 'Provider_Name', 'Provider_Address', 'Provider_City', 'Provider_State', 'Provider_Zip', 'Market', 'Total_Discharges', 'Average_Covered_Charges', 'Average_Total_Payments', 'Average_Medicare_Payments']
```

```python
tn['Cost_per_Discharge'] = (tn['Average_Total_Payments'] / tn['Total_Discharges'])
tn['Percentage_Covered_by_Medicare'] = (tn['Average_Medicare_Payments'] / tn['Average_Total_Payments'])
tn['Difference_Not_Covered'] = (tn['Average_Total_Payments'] - tn['Average_Medicare_Payments'])
```
